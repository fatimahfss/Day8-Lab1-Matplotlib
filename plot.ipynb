```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Q1: Create two charts using list and dataframe


```python
plt.plot([1,2,3,4])

x=pd.DataFrame({"name":['Fatimah','Deema','Kholoud'],"grade":[98,77,45]})
x.plot.bar(x="name",y="grade")
```

# Q2: Create pie chart with equal explode values.

```python
z=plt.pie(x=np.arange(5), explode=[0,0,0,0,0])
```

# Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
s=plt.scatter(x=np.arange(5),marker='o', y=np.arange(5), alpha=0.4, s=50);
```

# Q4: Add the following to your chart: 
*title placed in the left and has a red color
*x and y label with font size = 16

```python
s=plt.scatter(x=np.arange(5), y=np.arange(5), alpha=0.4, s=50);
plt.title("Pie chart" , loc='left', color='r')
plt.xlabel('xlable', fontsize=16)
plt.ylabel('ylable', fontsize=16)
```

# Q5: Create three line charts with Legends at left and colors

```python
x=[1,2,3]
y=[3,4,5]
z=[5,6,1]
plt.plot(x, color='blue')
plt.plot(y, color='yellow')
plt.plot(z, color='red')
plt.legend(['x','y','z'],loc ='upper left')
```

# Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 16)
#first
plt.subplot(2,2,1) #(number of rows, number of columns, number of plot)
plt.pie(z)
plt.title('Pie chart')

#second
plt.subplot(2,2,2)
plt.plot(x)
plt.plot(y)
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');
#third
plt.subplot(2,2,3)
plt.plot(x, color='blue')
plt.plot(y, color='yellow')
plt.plot(z, color='red')
plt.legend(['x','y','z'],loc ='upper left')
```
